```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
import wikipedia as wiki
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
# nltk.download('language')
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words,end="")
```

```python
text = wiki.page("covid-19").content
```

```python
word_tokens = word_tokenize(text,language='english')
print(word_tokens)
```

```python
filtered_word = [i for i in text if i not in stop_words]
unique_words = set(word_tokens)
set_of_unique_words_withought_stop_words = unique_words.difference(stop_words)
print(set_of_unique_words_withought_stop_words,end="")
print("*"*50)
# print(str(filtered_word).strip(),end="")
```

```python
# sentence tokenization
sentence_tokens = sent_tokenize(text,language='english')
filtered_sentence = [i for i in str(text).split(".") if i not in stop_words]
unique_sentence = set(sentence_tokens)
sentence_withought_stop_word = unique_sentence.difference(stop_words)
print(sentence_withought_stop_word)
print("*"*50)
# print(str(filtered_sentence).replace(" ",""),end="")
```
